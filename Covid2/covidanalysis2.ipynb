```python
import pandas as pd
```

```python
# Make a reference to the books.csv file path
csv_path = "WHO-COVID-19-global-data.csv"

# Import the books.csv file as a DataFrame
covid_df = pd.read_csv(csv_path, encoding="utf-8")
covid_df.head()
```

```python
covid_2_df = covid_df[["Date_reported", "Country", "New_cases", "Cumulative_cases",
                       "New_deaths", "Cumulative_deaths"]]
covid_2_df.head()
```

```python
covid_3_df = covid_2_df.rename(columns={'Date_reported': 'Date', 'New_cases': 'New Cases','Cumulative_cases':'Cumulative cases', 'New_deaths':'New deaths', 'Cumulative_deaths':'Cumulative deaths'})
covid_3_df.head()
```

```python
us_covid_df = covid_3_df.loc[(
    covid_3_df["Country"] == "United States of America")]
us_covid_df.head()
```

```python
us_covid_df.to_csv('US_covid_2.csv', index=False)
```
